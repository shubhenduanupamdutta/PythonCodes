## Getting data from PaySafe website

```python
!pip install webdriver-manager
# Importing all modules
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

URL = "https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors"
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
EMAIL_ID = "YOUR_EMAIL"
PASSWORD = "YOUR_PASSWORD"
```

```python
driver.get(url=URL)
```

```python
accept_cookies = driver.find_element(By.ID, "onetrust-accept-btn-handler")
accept_cookies.click()
```

```python
rank = []
major = []
degree = []
early_career = []
mid_career = []
high_meaning = []

def get_data_value(cell_row, cell_column):
    return cell_row[cell_column].text

there_is_next_page = True
while there_is_next_page:
    data_rows_on_page = driver.find_elements(By.CSS_SELECTOR, "table.data-table tbody tr")
    for data_row in data_rows_on_page:
        data_cells = data_row.find_elements(By.TAG_NAME, "td")
        rank.append(get_data_value(data_cells, 0))
        major.append(get_data_value(data_cells, 1))
        degree.append(get_data_value(data_cells, 2))
        early_career.append(get_data_value(data_cells, 3))
        mid_career.append(get_data_value(data_cells, 4))
        high_meaning.append(get_data_value(data_cells, 5))
    try:
        print("Checking next page button")
        driver.find_element(By.CSS_SELECTOR, ".pagination__btn.pagination__next-btn.pagination__btn--off")
        # Finds if next button is turned off
    except NoSuchElementException:
        # Next page button is not turned off
        next_button = driver.find_element(By.CSS_SELECTOR, ".pagination__btn.pagination__next-btn")
        next_button.click()
    else:
        print("Next button is disabled, coming out of loop.")
        there_is_next_page = False
```

```python
# Quit driver
driver.quit()
```

```python
df = pd.DataFrame({
    "Rank": rank,
    "Major": major,
    "Degree": degree,
    "Early_Career_Salary": early_career,
    "Mid_Career_Salary": mid_career,
    "High_Meaning": high_meaning,
})
df
```

```python
df.to_csv("UpdatedPaySafeData.csv", index=False)
```

```python
new_df = pd.read_csv("UpdatedPaySafeData.csv")
new_df
```

```python
new_df.drop("Rank", axis=1, inplace=True)
new_df.drop("Degree", axis=1, inplace=True)
```

```python
new_df["Early_Career_Salary"]= pd.to_numeric(new_df["Early_Career_Salary"].str.replace("$", "").str.replace(",", ""))
new_df.Mid_Career_Salary = pd.to_numeric(new_df.Mid_Career_Salary.str.replace("$", "").str.replace(",", ""))
```

```python
new_df.High_Meaning = pd.to_numeric(new_df.High_Meaning.str.replace("%", ""), errors="coerce")
new_df
```

```python
new_df.fillna(0)
```

```python
growth_in_salary = new_df["Mid_Career_Salary"] - new_df.Early_Career_Salary
new_df.insert(3, "Salary_Growth", growth_in_salary)
```
